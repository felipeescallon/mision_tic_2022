# Video Matrices:

```python
M=[[3,8,1],[-1,4,7],[3,2,0],[9,12,65]] 
for i in range(0,4): 
  for j in range(0,3): 
    print(M[i][j], end='\t')    
  print('\t')
```

# Video de Busqueda:

```python
personas=['123456789','987654321']
def buscar_id(id_s):
  for id in personas:
    if id==id_s:
      return id
  return None

id_encontrada1=buscar_id('123456789')
print(id_encontrada1)
print(buscar_id('123456789'))
print()
id_encontrada2=buscar_id('10')
print(id_encontrada2)
print(buscar_id('10'))

```
